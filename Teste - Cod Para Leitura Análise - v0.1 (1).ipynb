```python
!pip install pyautogui 
import pyautogui
import pandas as pd


df = pd.read_excel('Análise_Schuck_07-02(teste)tratado.xlsx')

#lendo os nomes das colunas(Onde ficam os cod's das lojas)
#Sempre verificar a quantidade de colunas(Melhoria: verificação automática).
#As colunas 0 e 1 são desconsideradas pois são as colunas de Dados de cabeçalho e Códigos
#Sempre verificar a quantidade de colunas(Melhoria: verificação automática)
#Intervalo de Colunas que serão lidas.

colunas = df.columns[2:41]

#Armazenando em uma nova lista os nomes das colunas.
#Os nomes das colunas são armazenados novamente em uma nova lista, pois, o que a variavel "colunas" retornou 
#Foi uma TUPLA que armazenou a lista de nome das colunas dificultando a leitura por index.
cln = []

#Armazenando em uma lista os nomes das colunas armazendas em "colunas" através de om laço de repetição.
for c in range(len(colunas)):
    
    c = colunas[c]
    
    cln.append(c)
    

#Usando o tamanho da lista "cln" onde é armazenado os nomes das colunas para definir a quantidade de repetições, conforme
#a quantidade de nomes que são a quantidade de lojas a serem digitadas:

for c in range(len(cln)):

    pyautogui.sleep(2)

    # Lendo a planilha
    planilha = pd.read_excel('Análise_Schuck_07-02(teste)tratado.xlsx')
    
    #Lendo os dados da coluna "Dados", onde há informações até a linha 7 que são utilizados no cabeçalho.
    valores_coluna = planilha['Dados'][:7]

    display(valores_coluna[1])
    
    #Inserindo dados no cabeçalho do pedido.
    
    #Empresa
    pyautogui.click(2104,185)
    #Pegando a empresa pela lista de nome de colunas armazenadas,onde será seguida a ordem de digitação conforme 
    #A ordem dos cod armazenados em "cln".
    pyautogui.typewrite(str(cln[c]))

    pyautogui.sleep(0.5)

    #Parceiro
    pyautogui.click(2104,258)
    pyautogui.typewrite(str(int(valores_coluna[1])))

    pyautogui.sleep(0.5)

    #TOP
    pyautogui.click(2104,329)
    pyautogui.typewrite(str(int(valores_coluna[2])))

    pyautogui.sleep(0.5)

    #TIPO NEGOCIAÇÃO
    pyautogui.click(2104,282)
    pyautogui.typewrite(str(int(valores_coluna[3])))

    pyautogui.sleep(0.5)

    #PREVISÃO DE ENTREGA
    pyautogui.click(2105,307)
    pyautogui.typewrite(str(int(valores_coluna[5])))

    pyautogui.sleep(0.5)

    #COMPRADOR
    pyautogui.click(2104,426)
    pyautogui.typewrite(str(int(valores_coluna[4])))
    
    #TRANSPORTADORA
    
    pyautogui.click(2104,497)
    pyautogui.typewrite(str(int(valores_coluna[6])))


    #CONFIRMA
    pyautogui.click(2024,117)

    # Carrega o arquivo Excel em um DataFrame do pandas
    df = pd.read_excel('Análise_Schuck_07-02(teste)tratado.xlsx')

    # Exibe o conteúdo do DataFrame
    print(df)


    pyautogui.sleep(3)
    
    #Posições de campos destinados aos códigos e quantidades.
    x_cod = 2706;
    y_cod = 234;

    x_qtd = 3310;
    y_qtd = 234;
    
    #Limite da tela durante a digitação onde o mause irá parar de se deslocar para baixo.
    x_cod_Limite_Tela = 2682
    y_cod_Limiye_Tela = 519
    
        
    x_qtd_Limite_Tela = 2800
    y_qtd_Limite_Tela = 519
     

    repeticao = 0
    
    #laço para ler coluna de código e coluna de quantidade até a linha 16.
    
    for index, row in df.iterrows():
        
        #Dentro deste if se a repetição for menor que 16, que é a repetição onde o mouse atinge o limite da tela
        #as coordenadas sempre receberão o valore de 20 para que dessa forma seja deslocada para o campo abaixo na 
        #próxima repetição.
        
        if repeticao < 16:
            
            #Verificando se a linha da coluna é vazia ou igual a 0 onde se não for nenhum dos casos os campos serão preenchidos:
            if not pd.isnull(row[cln[c]]) and row[cln[c]] != 0 and row[cln[c]] != '':
        
        
                #Inserindo Código
                pyautogui.click(x_cod, y_cod)
                #Neste cod o programa faz a leitura de toda coluna "Código" cada linha a cada repetição.
                pyautogui.typewrite(row['Código'])


                #Inserindo Quantidade
                pyautogui.click(x_qtd, y_qtd)
                pyautogui.sleep(1)
                #Na linha abaixo o programa pega o nome da coluna que deve ler de forma automatica
                pyautogui.typewrite(str(int(row[cln[c]])))


                pyautogui.sleep(2)

                
                #Confirmar
                pyautogui.click(2868,185)
                pyautogui.click(2868,185)
                pyautogui.sleep(5)

                #Novo Item
                pyautogui.click(2820,184)
                pyautogui.click(2820,184)
                pyautogui.sleep(1)
                
                #Soma dos valores nas coordenadas para que que o mouse realize o deslocamento para o próximo 
                #campo na proxima repetição.
                
                y_cod += 20;
                y_qtd += 20;
                
                repeticao += 1
                
        else: 
            
            
            if not pd.isnull(row[cln[c]]) and row[cln[c]] != 0 and row[cln[c]] != '':

                    #Inserindo Código
                    pyautogui.click(x_cod_Limite_Tela, y_cod_Limiye_Tela)
                    pyautogui.typewrite(row['Código'])


                    #Inserindo Quantidade
                    pyautogui.click(x_qtd_Limite_Tela, y_qtd_Limite_Tela)
                    pyautogui.sleep(1)
                    pyautogui.typewrite(str(int(row[cln[c]])))

                    pyautogui.sleep(2)

                    #Confirmar Item
                    pyautogui.click(2868,185)
                    pyautogui.click(2868,185)
                    pyautogui.sleep(5)

                    #Novo Item
                    pyautogui.click(2820,184)
                    pyautogui.click(2820,184)
                    pyautogui.sleep(1)

       
    #Confirmar Pedido
    #pyautogui.click(2168,116)
    #pyautogui.click(2168,116)
    
    pyautogui.sleep(3)
    
    #Novo pedido
    pyautogui.click(1970,117)
    pyautogui.click(1970,117)
```
